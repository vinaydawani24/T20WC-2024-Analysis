```python
import pandas as pd
import json
```

```python
# Correct way to load the JSON file with UTF-8 encoding
with open("data/matchResults.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df_match = pd.DataFrame(data)
df_match.head(5)
```

```python
match_id_dict = {}
for index,row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    match_id_dict[key1] = row['scorecard']
    match_id_dict[key2] = row['scorecard']

match_id_dict
```

```python
df_match.shape
```

```python
df_match.to_csv('data/match_summary_csv',index=False)
```

```python
# Correct way to load the JSON file with UTF-8 encoding
with open("data/battingSummary.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df_batting = pd.DataFrame(data)
df_batting.head(5)
```

```python
df_batting["out/not_out"] = df_batting["dismissal"].apply(lambda x: "not out" if x=="not out" else "out")
```

```python
df_batting.head(5)
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
df_batting.head(5)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_id_dict)

df_batting.head(5)
```

```python
df_batting.to_csv('data/batting_summary_csv',index=False)
```
